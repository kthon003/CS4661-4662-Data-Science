```python
from  keras.wrappers.scikit_learn  import  KerasClassifier
from  sklearn.model_selection  import  GridSearchCV
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
import numpy as np
from sklearn.metrics import accuracy_score
from keras.utils import np_utils
```

```python
from keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)
```

```python
print (y_train.shape)
print (y_train[:10,:])
```

```python
input_size = 784
hidden_neurons = 100
out_size = 10
```

```python
    def model_creator():

        model = Sequential()

        model.add(Dense(hidden_neurons, input_dim = input_size))
        model.add(Activation('sigmoid'))
        model.add(Dense(out_size, input_dim = hidden_neurons))
        model.add(Activation('softmax'))

        model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

        return model
```

```python
seed = 2 
np.random.seed(seed)
```

```python
model = KerasClassifier(build_fn = model_creator, verbose=2)
```

```python
batch_size = [30 , 50 , 100 ]
epochs = [10 , 15 , 20]
param_grid = dict(batch_size=batch_size, epochs=epochs)
```

```python
#4
grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = 10, scoring = 'neg_log_loss')
```

```python
grid.fit(X_train, y_train)
```

```python
#7
best_model = grid.best_estimator_.model
```

```python
grid.best_params_
```

```python
#8
batch_size = [50]
epochs = [20]
param_grid = dict(batch_size=batch_size, epochs=epochs)
grid_test = GridSearchCV(estimator = model, param_grid = param_grid, cv = 10, scoring = 'neg_log_loss')
```

```python
grid_test.fit(X_test, y_test)
```

```python
best_model_test = grid_test.best_estimator_.model
```

```python
y_predict = grid.best_estimator_.model.predict(X_test)
```

```python
score = grid.best_estimator_.model.evaluate(X_test, y_test, verbose = 1)
```
